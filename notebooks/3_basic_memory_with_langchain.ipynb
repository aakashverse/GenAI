##### Adding Memory to AI Models

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
res = llm.invoke("Hello, My name is Akash Yadav")
res.content
```

```python
res = llm.invoke("What is my Name?")
res.content
```

```python
prompts = [
    {"role":"user", "content":"My name is Akash Yadav."},
    {"role":"ai", "content":"Hello Akash Yadav! It's nice to meet you. How can I help you today?"},
    {"role":"user", "content":"What is my Name?"}
]
```

```python
res = llm.invoke(prompts)
res.content
```

##### Manga History

```python
history = []
while True:
    query = input("User: ")

    if query.lower() in ["exit", "quit", "bye"]:
        print("Good Bye :)")
        break

    history.append({"role":"user", "content":query})
    print("User: ", query)

    res = llm.invoke(history)
    history.append({"role":"ai", "content":res.content})
    print("AI: ", res.content, "\n")
```
