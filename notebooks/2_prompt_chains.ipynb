##### Prompt Template - Dynamic Prompt

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
out = StrOutputParser()  
```

```python
# My Self Chain Node
def transform_case(result:str):
    return result.upper()
```

```python
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a Translator & translate the input in {language}"},
    {"role": "user", "content": "{query}"},
])
prompts
```

```python
# objects that can be invoked are called as 'Runables', here prompts & llm are runnables
# Chaining => prompts.invoke() ke output ko llm.invoke() ke as input bhej diya
res = llm.invoke(prompts.invoke({"language":"Hindi", "query":"I love You"}))
res.content
# but this is not the correct way to make chains
```

##### prompts ka output -> as input to llm -> llm ka output -> as input to out, here chains is also a runable

```python
chains = prompts | llm | out | transform_case
```

```python
res = chains.invoke({"language":"Hinglish", "query":"I love LangChain"}) 
print(res)
```
