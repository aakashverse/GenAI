##### Tools are callable functions with well-defined inputs and outputs that get passed to a chat model. The model decides when to invoke a tool based on the conversation context, and what input arguments to provide.

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.tools import tool
from langchain.agents import create_agent
```

```python
@tool
def add_numbers(a:int, b:int):
    """
      it will return the sum of two numbers

      Args:
        a: Number One
        b: Number Two
    """
    return a + b

@tool
def multiply_numbers(a:int, b:int):
    """
      it will return the product of two numbers

      Args:
        a: Number One
        b: Number Two
    """
    return a * b
```

```python
add_numbers.invoke({"a":12, "b":23})
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model = "gemini-2.5-flash-lite")
```

```python
agent = create_agent(
    model=llm,
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a math teacher, and always use tool for calculation."
)
```

```python
response = agent.invoke({"messages":[{"role": "user", "content":"what is 2 + 30 * 10?"}]})
for res in response['messages']:
    print(res)
    print("\n")

# flow of Agent -> 1) Reasoning & thinking , 2) Action(calling the Tool) , 3) Observing , 4) Generate the Response
```
