##### Streaming LLM Response - [displaying/getting data in chunk form]

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite", streaming=True)
```

```python
question ="can you explain me GEN_AI in short?"
```

```python
res = llm.stream(question)  # stream provides a generator (stream object) & after execution it clean its memory
for chunk in res:
    print(chunk.content, end="")
```

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="openai/gpt-oss-120b", streaming=True)
```

```python
question = "can you explain Gen-AI?"
res = llm.stream(question)
for chunk in res:
    print(chunk.content, end="")
```
