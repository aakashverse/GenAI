```python
from pydantic import BaseModel, Field
from typing import List, Annotated
```

```python
class User(BaseModel):
    name : str = "",
    age : int
```

```python
user = User(name="Akash", age=22)
```

```python
### Field validation
```

```python
class Product(BaseModel):
    title : str = Field(max_length=10)
    desc : str = Field(min_length=20)
    price : int = Field(gt=0, le=100)
```

```python
product = Product(title="Samsung", desc="This is an amazing smartphone from samsung.", price=90)
product2 = Product(title="onePlus", desc="This is an amazing smartphone from onePlus.", price=80)
```

```python
class Order(BaseModel):
    id : str = Field(min_length=6, max_length=12)
    product : Product
```

```python
order = Order(id="j21369", product=product)
```

```python
order.product.price
```

```python
class Carts(BaseModel):
    id : str = Field(min_length=6, max_length=12)
    items : List[Product]
    comment : str = Field(
        default = "No Comment"
    )
```

```python
cart = Carts(id="mv2e99", items=[product, product2], comment="This is my cart")
```

```python
cart
```

```python
# Annotated - can build custom validator
commonValidator = Annotated[List[Product], Field(default=[])]
```

```python
class Carts(BaseModel):
    id : str = Field(min_length=6, max_length=12)
    items : commonValidator
    comment : str = Field(
        default = "No Comment"
    )
```

```python
class Order(BaseModel):
    products: commonValidator
```

```python
ord = Order(products=[product, product2])
```
