##### Connect with OpenAI model

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
res = llm.invoke("How to find the largest number from a list in python ?")
res.content
```

##### Connect with Anthropic

```python
from langchain_anthropic import ChatAnthropic
llm = ChatAnthropic(model="claude-sonnet-4-5")
```

```python
res = llm.invoke("Who won IPL 2025 ?")
print(res.content)
```

##### Connect wih Google Gemini model

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
res = llm.invoke("What is GenAI? Explain in short ?")
print(res.content)
```

##### Static Prompts

```python
prompts = [
    ("system", "You are a cricket Expert."),  ## type: 'system', 'ai', 'user'
    ("user", "Who won WIPL 2024 ?")
]

res = llm.invoke(prompts)
print(res.content)
```
